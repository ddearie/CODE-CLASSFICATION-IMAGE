```python
import os
import cv2

def convert_images_to_png(input_dir, output_dir):
    if not os.path.exists(input_dir):
        print(f"Thư mục {input_dir} không tồn tại.")
        return

    os.makedirs(output_dir, exist_ok=True)
    for filename in os.listdir(input_dir):
        input_path = os.path.join(input_dir, filename)
        image = cv2.imread(input_path)
        if image is not None:
            output_filename = os.path.splitext(filename)[0] + ".png"
            output_path = os.path.join(output_dir, output_filename)
            cv2.imwrite(output_path, image, [int(cv2.IMWRITE_PNG_COMPRESSION), 0])

            print(f"Chuyển đổi thành công: {input_path} -> {output_path}")
        else:
            print(f"Lỗi khi đọc ảnh: {input_path}")


train_dir = '/content/data/train/fractured'
train_output_dir = '/content/train/fractured_PNG'
train_dir_z = '/content/data/train/non_fractured'
train_output_dir_z = '/content/train/non_fractured_PNG'

val_dir = '/content/data/val/fractured'
val_output_dir = '/content/val/fractured_PNG'
val_dir_z = '/content/data/val/non_fractured'
val_output_dir_z = '/content/val/non_fractured_PNG'


convert_images_to_png(train_dir_z, train_output_dir_z)
convert_images_to_png(val_dir_z, val_output_dir_z)

convert_images_to_png(train_dir, train_output_dir)
convert_images_to_png(val_dir, val_output_dir)
```

```python
#import libraries for pre-processing
import numpy as np
import cv2
import os

def roı_clahe_pre_process(folder,new_folder):
    if not os.path.exists(folder):
        print(f"Thư mục {folder} không tồn tại.")
        return

    os.makedirs(new_folder, exist_ok=True)

    for filename in os.listdir(folder):
        input_path = os.path.join(folder, filename)

        img = cv2.imread(input_path, 0) #read image from directory
        thresh = cv2.threshold(img, 0, 255, cv2.THRESH_BINARY + cv2.THRESH_OTSU)[1] #apply binary and otsu threshold
        x,y,w,h = cv2.boundingRect(thresh) #determine boundary of ROI
        x, y, w, h = x, y, w+20, h+20 #+20 pixels tolerans for boundaries
        img = img[y:y+h, x:x+w] #crop original image with the help of boundary

        clahe = cv2.createCLAHE(clipLimit=2.0, tileGridSize=(8, 8)) #determine clahe values
        img = clahe.apply(img) #apply clahe transform on image

        new_path=os.path.join(new_folder, filename) #determine new path for save to image
        cv2.imwrite(new_path, img) #save output to new paths

folder="/content/train/fractured_PNG"
new_folder="/content/train_1/fractured_PNG"

folder_1="/content/train/non_fractured_PNG"
new_folder_1="/content/train_1/non_fractured_PNG"

folder_2="/content/val/fractured_PNG"
new_folder_2="/content/val_1/non_fractured_PNG"

folder_3="/content/val/non_fractured_PNG"
new_folder_3="/content/val_1/non_fractured_PNG"

#run funnction with folder paths
roı_clahe_pre_process(folder,new_folder)
roı_clahe_pre_process(folder_1,new_folder_1)
roı_clahe_pre_process(folder_2,new_folder_2)
roı_clahe_pre_process(folder_3,new_folder_3)
```
